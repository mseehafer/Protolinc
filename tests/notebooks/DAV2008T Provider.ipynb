### DAV2008T

```python
import os
import numpy as np
import pandas as pd
from pyprotolinc import _DEFAULT_TABLES_PATH

base_directory = os.path.join(_DEFAULT_TABLES_PATH, "Germany_Endowments_DAV2008T")
base_rates_path = os.path.abspath(os.path.join(base_directory, "Germany_Endowments_DAV2008T.csv"))
base_rates_path
```

```python
tab_DAV2008T = pd.read_csv(base_rates_path, header=[0, 3, 4])
tab_DAV2008T.head()
```

```python
cols = tab_DAV2008T.columns
cols
```

```python
# forward fill headers
filled_headers = [[], [], []]

for row_count, ind_row in enumerate(cols):
    for l in range(len(filled_headers)):
        if row_count > 0 and ind_row[l].startswith("Unnamed:"):
            filled_headers[l].append(filled_headers[l][-1])
        else:
            filled_headers[l].append(ind_row[l])
filled_headers
```

```python
new_headers = pd.MultiIndex.from_arrays(filled_headers, names=('GENDER_IND', 'PRUDENCE_RELATED', "SMOKER_RELATED"))
new_headers
```

```python
tab_DAV2008T.columns = new_headers
tab_DAV2008T.head()
```

```python
#headers_males = new_headers[4:10]
#headers_females = new_headers[14:]
#tab_DAV2008T_males = tab_DAV2008T[headers_males]
#tab_DAV2008T_females = tab_DAV2008T[headers_females]

#tab_DAV2008T_females.head()
```

```python
tab_DAV2008T_transf = tab_DAV2008T[new_headers[list(range(4, 10)) + list(range(14, len(new_headers)))]]

#    .unstack().reset_index().drop("level_3", axis=1)
tab_DAV2008T_transf
```

```python
from pyprotolinc.models.risk_factors import Gender, SmokerStatus, Age
from pyprotolinc.assumptions.providers import AssumptionType

remapped_headers = []

gender_map = {
    'DAV 2008T R/NR Männer': Gender.M,
    'DAV 2008T R/NR Frauen': Gender.F
}

assumption_type_map = {
    'Sterblichkeit 2. Ordnung': AssumptionType.BE,
    'Sterblichkeit 1. Ordnung': AssumptionType.RES,
}

smoker_map = {
   'Aggregat': SmokerStatus.A,
    ' Nichtraucher': SmokerStatus.N,
    'Raucher': SmokerStatus.S,
    'Aggregat (Zuschlag 34%)': SmokerStatus.A,
    'Nichtraucher (Zuschlag 45%)':SmokerStatus.N,
    'Raucher (Zuschlag 45%)': SmokerStatus.S
}

for col in tab_DAV2008T_transf.columns:
    remapped_headers.append((gender_map[col[0]], assumption_type_map[col[1]], smoker_map[col[2]]
    ))

tab_DAV2008T_transf.columns = remapped_headers


tab_DAV2008T_transf
```

```python
# use a 3D array to store the rates with dimensions
# Age x Gender x Smoker

be_rates = np.zeros((len(tab_DAV2008T), 2, 3))
res_rates = np.zeros((len(tab_DAV2008T), 2, 3))

for g, at, ss in remapped_headers:
    if at == AssumptionType.BE:
        be_rates[:, g, ss] = tab_DAV2008T_transf[(g, at, ss)]
    elif at == AssumptionType.RES:
        res_rates[:, g, ss] = tab_DAV2008T_transf[(g, at, ss)]
    else:
        raise Exception("Unknown assumption type")
be_rates
```

```python

from pyprotolinc.assumptions.providers import StandardRatesProvider
```

```python
provider_be = StandardRatesProvider(be_rates,
                                    (Age, Gender, SmokerStatus),
                                     offsets=(0, 0, 0))

provider_res = StandardRatesProvider(res_rates,
                                    (Age, Gender, SmokerStatus),
                                     offsets=(0, 0, 0))
```

```python
ages = np.array([10, 20, 30], dtype=np.int32)
smokerstatus = np.array([SmokerStatus.S, SmokerStatus.N, SmokerStatus.A], dtype=np.int32)
genders = np.array([Gender.M, Gender.M, Gender.F], dtype=np.int32)


provider_be.get_rates(age=ages, smokerstatus=smokerstatus, gender=genders)
```

```python
provider_res.get_rates(age=ages, smokerstatus=smokerstatus, gender=genders)
```
