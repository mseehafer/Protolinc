```python
import cProfile
import pstats, math
import io
import pandas as pd
 
pr = cProfile.Profile()
pr.enable()
project_cashflows_cli()
pr.disable()
 
result = io.StringIO()
pstats.Stats(pr,stream=result).print_stats()
result=result.getvalue()
# chop the string into a csv-like buffer
result='ncalls'+result.split('ncalls')[-1]
result='\n'.join([','.join(line.rstrip().split(None,5)) for line in result.split('\n')])
# save it to disk
 
with open('profile_test.csv', 'w+') as f:
    #f=open(result.rsplit('.')[0]+'.csv','w')
    f.write(result)
    f.close()
```

```python
df = pd.read_csv(io.StringIO(result))
```

```python
df_filtered = df[(df["filename:lineno(function)"].str.contains("protolinc")) &
                 (df.cumtime > 0)].copy()
df_filtered.head()
```

```python
# common prefix
start_pos = df_filtered["filename:lineno(function)"].str.find("protolinc").unique()[0]
df_filtered["filename:lineno(function)"] = df_filtered["filename:lineno(function)"].str.slice(start_pos)
```

```python
import numpy as np
df_filtered.index = np.arange(1, 1 + len(df_filtered))
```
